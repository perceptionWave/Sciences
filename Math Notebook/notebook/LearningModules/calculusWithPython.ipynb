```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
import sympy as sy
#%matplotlib inline
```

```python
# symbols and init_printing
init_printing(use_unicode=True)
x, y, z, t = symbols('x y z t')
k, m, n = symbols('k m n', integer=True)
f, g, h = symbols('f g h', cls=Function)
foo = diff(sin(x)*exp(x), x)
print(foo)
```

```python
def f_1(x): return x**2-2*x+1
def g_1(x): return x**2-5*x+4

# plot f1(x) and g1(x) on the interval[2,4]
x = np.linspace(2, 4, 100) # generates 100 points linearly spaced between and including 2 and 4
plt.plot(x, f_1(x), '-b', label='f1(x)')
plt.plot(x, g_1(x), '-r', label='g1(x)')

# fill in the area between two curves f(x) and g(x), alpha is transparency level
plt.fill_between(x, f_1(x), g_1(x), color='skyblue', alpha=0.5)

# add titles, lables and legends
plt.title("The area between two curves")
plt.xlabel("x")
plt.ylabel("Y")
plt.xticks([2,4],['2','4'])
plt.yticks()
plt.legend()
plt.show()
```

```python
# compute the area by the definite integral formula
x = sy.Symbol('x') # x is the variable 
area = sy.integrate(f_1(x)-g_1(x),(x,2,4)) # f(x) is the upper curve, while g is lower, so we have to put f(x)-g(x)
print("Area = " + str(area))
```

```python
def f_2(x): return x**2
def g_2(x): return 4*x-x**2

x = np.linspace(0,3,100)
plt.plot(x,f_2(x), '-b', label='f2(x)')
plt.plot(x,g_2(x), '-r', label='g2(x)')

plt.fill_between(x, f_2(x),g_2(x), color='skyblue',alpha=0.5)

plt.title("The area between two curves")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.xticks([0,2],['0','2'])
plt.yticks([])
plt.show()
```

```python
x = sy.Symbol('x')
result = sy.integrate((g_2(x)-f_2(x)), (x, 0, 2)) # f2(x) is the lower curve, while g is higher, so we have to put g(x)-f(x) 
# or, result = sy.integrate(diff(x), (x, 0, 2))

print("Area = " + str(result))
```

```python
def f_2(x): return x**2
def g_2(x): return 4*x-x**2
def diff(x): return 4*x-2*x**2

x = np.linspace(0,2,100)
plt.plot(x,f_2(x), '-b', label='f2(x)')
plt.plot(x,g_2(x), '-r', label='g2(x)')
plt.plot(x,diff(x), '-g', label='diff(x)')

plt.axhline(color='black') # draw the x-axis
plt.fill_between(x, f_2(x),g_2(x), color='skyblue',alpha=0.5)
plt.fill_between(x, diff(x), color='olive', alpha=0.5) # the area bounded by the graph of g(x)-f(x)

plt.title("The area between two curves")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.xticks([0,2],['0','2'])
plt.yticks([])
plt.show()
```
