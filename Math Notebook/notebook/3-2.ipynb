<h1>Question 13</h1>
<p>The curve y=x/(1 + x2) is called a serpentine. \nFind an equation of the tangent line to this curve at the point\n(2, 0.40).</p>

```python
def s32q13(x): return(x/(1+x**2))

q = diff(s32q13(x),x)
print("f'(x) = " + str(q))

q = simplify(diff(s32q13(x),x))
print("f'(x) = " + str(q))

def derq13(x): return((1 - x**2)/(x**4 + 2*x**2 + 1))
m = derq13(2)
print("f'(2) = "+ str(m))

def q13tan(x): return(m*(x-2)+0.4)

print('Answer')
q = simplify(m*(x-2)+0.4)
print('y = ' + str(q))
```

```python
i = np.linspace(-2,6, 500)
plt.title('y=x/(1+x^2')
plt.plot(i, s32q13(i), 'r', label="f(x)")
plt.plot(i, derq13(i), 'blue', label="f'(x)")
plt.plot(i, q13tan(i), 'black', label='tan(2)')
plt.grid()
plt.legend()
plt.ylim(-1,1.2)
plt.show()
```
